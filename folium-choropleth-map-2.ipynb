```python
import folium
import requests as r
import json
import pandas as pd
from IPython.display import HTML

data_dict_ = {'zip_code': [90221],
             'average_score': [99.492847],
             'total_facilities': [123]}
df_data = pd.DataFrame(data_dict_)
print(df_data)

ca_json = r.get('http://bit.ly/ca_zip')
print(ca_json)

cali_map = folium.Map(location=[33.895847, -118.220070], zoom_start=12)
folium.Choropleth(geo_data=ca_json.json()['features'][0], 
                     fill_color='OrRd',
                     fill_opacity=0.2,
                     line_opacity=0.5,
                     data = df_data,
                     columns=['zip_code','average_score'],
                     key_on='feature.properties.ZCTA5CE10').add_to(cali_map)

cali_map.save('plot_data_2.html')
HTML('<iframe src=plot_data_2.html width=800 height=500></iframe>')
```
