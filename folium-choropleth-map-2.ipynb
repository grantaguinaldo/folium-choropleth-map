```python
#http://comet.lehman.cuny.edu/owen/teaching/datasci/choroplethLab.html
json_data = r.get('http://comet.lehman.cuny.edu/owen/teaching/datasci/schoolDistricts.json')
```

```python
geo_data_json = json_data.json()
```

```python
#https://www.kdnuggets.com/2018/09/visualising-geospatial-data-python-folium.html

school_map = folium.Map(location=[40.75, -74.125], tiles='Mapbox Bright', zoom_start=11)
folium.Choropleth(geo_data=geo_data_json, 
                     fill_opacity=0.2,
                     line_opacity=0.5).add_to(school_map)
school_map.save('plot_data.html')
HTML('<iframe src=plot_data.html width=800 height=700></iframe>')
```
